```python
# Author: Adam Fletcher
# Date: 12/13/2022
# Purpose: To simulate through animation the elliptical orbits of the planets 
#          around the sun.

import matplotlib.pyplot as plt
from matplotlib import animation, rc
rc('animation', html='jshtml')
from matplotlib.animation import FuncAnimation

import matplotlib
matplotlib.rcParams['animation.embed_limit'] = 2**128

from IPython.display import HTML
#from os import major

         ####### Variable Dictionary #######
G           = 6.67e-11                  # Gravitational Const
                                        # Masses (kg)
 
Ms          = 2.0e30                    # sun 
Mmer        = 3.30104e23                # mercury
Mv          = 4.87e24                   # venus
Me          = 5.972e24                  # earth        
Mm          = 6.39e23                   # mars
Mj          = 1898.0e24                 # jupiter
Msat        = 568.0e24                  # saturn
Mu          = 86.8e24                   # uranus
Mn          = 1.024e26                  # neptune

AU          = 1.5e11                    # Astronomical Unit (meters)
daysec      = 24.0*60*60                # time (seconds)

                                        # (m/s)
mer_ap_v      = 47900                   # mercury velocity at aphelion
ven_ap_v      = 35020                   # venus velocity at aphelion
e_ap_v      = 29290                     # earth velocity at aphelion
m_ap_v      = 21970                     # mars velocity at aphelion
j_ap_v      = 13100                     # jupiter velocity at aphelion
sat_ap_v    = 9700                      # saturn velocity at aphelion
u_ap_v      = 6800                      # uranus velocity at aphelion
n_ap_v      = 5400                      # neptune velocity at aphelion


gravconst_mer = G*Mmer*Ms               # force of gravity on each planet
gravconst_ven = G*Mv*Ms
gravconst_e = G*Me*Ms
gravconst_m = G*Mm*Ms
gravconst_j = G*Mj*Ms
gravconst_sat = G*Msat*Ms
gravconst_u = G*Mu*Ms
gravconst_n = G*Mn*Ms



######## initial conditions ########

# mercury
xmer,ymer,zmer    = 0.4*AU,0,0
xvmer,yvmer,zvmer = 0,mer_ap_v,0

# venus
xven,yven,zven    = 0.7*AU,0,0
xvven,yvven,zvven = 0,ven_ap_v,0

# earth
xe,ye,ze    = 1.0167*AU,0,0
xve,yve,zve = 0,e_ap_v,0

# mars
xm,ym,zm    = 1.666*AU,0,0
xvm,yvm,zvm = 0,m_ap_v,0

# jupiter
xj,yj,zj    = 5.2*AU,0,0
xvj,yvj,zvj = 0,j_ap_v,0

# saturn
xsat,ysat,zsat    = 9.5*AU,0,0
xvsat,yvsat,zvsat = 0,sat_ap_v,0

# uranus
xu,yu,zu    = 19.8*AU,0,0
xvu,yvu,zvu = 0,u_ap_v,0

# neptune
xn,yn,zn    = 30.0*AU,0,0
xvn,yvn,zvn = 0,n_ap_v,0

# sun
xs,ys,zs    = 0,0,0
xvs,yvs,zvs = 0,0,0

t           = 0.0
dt          = 1*daysec                    # change in time with each new frame

xmerlist,ymerlist,zmerlist = [],[],[]
xvlist,yvlist,zvlist = [],[],[]
xelist,yelist,zelist = [],[],[]
xslist,yslist,zslist = [],[],[]
xmlist,ymlist,zmlist = [],[],[]
xjlist,yjlist,zjlist = [],[],[]
xsatlist,ysatlist,zsatlist = [],[],[]
xulist,yulist,zulist = [],[],[]
xnlist,ynlist,znlist = [],[],[]


####### begin simulation #######

while t<5*365*daysec:

    ################ mercury #############
    # compute G force on mercury
    
    rx,ry,rz = xmer - xs, ymer - ys, zmer - zs
    modr3_mer = (rx**2+ry**2+rz**2)**1.5
    fx_mer = -gravconst_mer*rx/modr3_mer
    fy_mer = -gravconst_mer*ry/modr3_mer
    fz_mer = -gravconst_mer*rz/modr3_mer
    
    # Apply Newton's Second Law  F = ma -> a = F/m
    xvmer += fx_mer*dt/Mmer
    yvmer += fy_mer*dt/Mmer
    zvmer += fz_mer*dt/Mmer
    
    # update position
    xmer += xvmer*dt
    ymer += yvmer*dt 
    zmer += zvmer*dt
    
    # save the position in list
    xmerlist.append(xmer)
    ymerlist.append(ymer)
    zmerlist.append(zmer)

    ################ venus #############
    # compute G force on venus
    
    rx,ry,rz = xven - xs, yven - ys, zven - zs
    modr3_ven = (rx**2+ry**2+rz**2)**1.5
    fx_ven = -gravconst_ven*rx/modr3_ven
    fy_ven = -gravconst_ven*ry/modr3_ven
    fz_ven = -gravconst_ven*rz/modr3_ven
    
    # Apply Newton's Second Law  F = ma -> a = F/m
    xvven += fx_ven*dt/Mv
    yvven += fy_ven*dt/Mv
    zvven += fz_ven*dt/Mv
    
    # update position
    xven += xvven*dt
    yven += yvven*dt 
    zven += zvven*dt
    
    # save the position in list
    xvlist.append(xven)
    yvlist.append(yven)
    zvlist.append(zven)

    ################ earth #############
    # compute G force on earth
    
    rx,ry,rz = xe - xs, ye - ys, ze - zs
    modr3_e = (rx**2+ry**2+rz**2)**1.5
    fx_e = -gravconst_e*rx/modr3_e
    fy_e = -gravconst_e*ry/modr3_e
    fz_e = -gravconst_e*rz/modr3_e
    
    # Apply Newton's Second Law  F = ma -> a = F/m
    xve += fx_e*dt/Me
    yve += fy_e*dt/Me
    zve += fz_e*dt/Me
    
    # update position
    xe += xve*dt
    ye += yve*dt 
    ze += zve*dt
    
    # save the position in list
    xelist.append(xe)
    yelist.append(ye)
    zelist.append(ze)
    
    ################ Mars ##############
    # compute G force on mars

    rx_m,ry_m,rz_m = xm - xs, ym - ys, zm - zs
    modr3_m = (rx_m**2+ry_m**2+rz_m**2)**1.5
    fx_m = -gravconst_m*rx_m/modr3_m
    fy_m = -gravconst_m*ry_m/modr3_m
    fz_m = -gravconst_m*rz_m/modr3_m
    
    # Apply Newton's Second Law  F = ma -> a = F/m
    xvm += fx_m*dt/Mm
    yvm += fy_m*dt/Mm
    zvm += fz_m*dt/Mm
    
    # update position
    xm += xvm*dt
    ym += yvm*dt
    zm += zvm*dt
    
    # add to list
    xmlist.append(xm)
    ymlist.append(ym)
    zmlist.append(zm)
    
    ################ jupiter #############
    # compute G force on jupiter
    
    rx,ry,rz = xj - xs, yj - ys, zj - zs
    modr3_j = (rx**2+ry**2+rz**2)**1.5
    fx_j = -gravconst_j*rx/modr3_j
    fy_j = -gravconst_j*ry/modr3_j
    fz_j = -gravconst_j*rz/modr3_j
    
    # Apply Newton's Second Law  F = ma -> a = F/m
    xvj += fx_j*dt/Mj
    yvj += fy_j*dt/Mj
    zvj += fz_j*dt/Mj
    
    # update position
    xj += xvj*dt
    yj += yvj*dt 
    zj += zvj*dt
    
    # save the position in list
    xjlist.append(xj)
    yjlist.append(yj)
    zjlist.append(zj)

    ################ saturn #############
    # compute G force on saturn
    
    rx,ry,rz = xsat - xs, ysat - ys, zsat - zs
    modr3_sat = (rx**2+ry**2+rz**2)**1.5
    fx_sat = -gravconst_sat*rx/modr3_sat
    fy_sat = -gravconst_sat*ry/modr3_sat
    fz_sat = -gravconst_sat*rz/modr3_sat
    
    # Apply Newton's Second Law  F = ma -> a = F/m
    xvsat += fx_sat*dt/Msat
    yvsat += fy_sat*dt/Msat
    zvsat += fz_sat*dt/Msat
    
    # update position
    xsat += xvsat*dt
    ysat += yvsat*dt 
    zsat += zvsat*dt
    
    # save the position in list
    xsatlist.append(xsat)
    ysatlist.append(ysat)
    zsatlist.append(zsat)

    ################ uranus #############
    # compute G force on uranus
    
    rx,ry,rz = xu - xs, yu - ys, zu - zs
    modr3_u = (rx**2+ry**2+rz**2)**1.5
    fx_u = -gravconst_u*rx/modr3_u
    fy_u = -gravconst_u*ry/modr3_u
    fz_u = -gravconst_u*rz/modr3_u
    
    # Apply Newton's Second Law  F = ma -> a = F/m
    xvu += fx_u*dt/Mu
    yvu += fy_u*dt/Mu
    zvu += fz_u*dt/Mu
    
    # update position
    xu += xvu*dt
    yu += yvu*dt 
    zu += zvu*dt
    
    # save the position in list
    xulist.append(xu)
    yulist.append(yu)
    zulist.append(zu)

    ################ neptune #############
    # compute G force on neptune
    
    rx,ry,rz = xn - xs, yn - ys, zn - zs
    modr3_n = (rx**2+ry**2+rz**2)**1.5
    fx_n = -gravconst_n*rx/modr3_n
    fy_n = -gravconst_n*ry/modr3_n
    fz_n = -gravconst_n*rz/modr3_n
    
    # Apply Newton's Second Law  F = ma -> a = F/m
    xvn += fx_n*dt/Mn
    yvn += fy_n*dt/Mn
    zvn += fz_n*dt/Mn
    
    # update position
    xn += xvn*dt
    yn += yvn*dt 
    zn += zvn*dt
    
    # save the position in list
    xnlist.append(xn)
    ynlist.append(yn)
    znlist.append(zn)
    
    ################ the sun ###########
    # Apply Newton's Second Law  F = ma -> a = F/m
    xvs += -(fx_e+fx_m)*dt/Ms
    yvs += -(fy_e+fy_m)*dt/Ms
    zvs += -(fz_e+fz_m)*dt/Ms
    
    # update position
    xs += xvs*dt
    ys += yvs*dt 
    zs += zvs*dt
    xslist.append(xs)
    yslist.append(ys)
    zslist.append(zs)
    
    # update dt
    t +=dt

#print(xelist,yelist)

####### plot the solar system ####### 


fig, ax = plt.subplots(figsize=(10,10))
ax.set_aspect('equal')
ax.grid()


line_mer,     = ax.plot([],[],'-g',lw=1)
point_mer,    = ax.plot([AU], [0], marker="o", markersize=2, markeredgecolor="brown", markerfacecolor="brown")
text_mer      = ax.text(AU,0,'Mercury')

line_ven,     = ax.plot([],[],'-g',lw=1)
point_ven,    = ax.plot([AU], [0], marker="o", markersize=2, markeredgecolor="green", markerfacecolor="green")
text_ven      = ax.text(AU,0,'Venus')

line_e,     = ax.plot([],[],'-g',lw=1)
point_e,    = ax.plot([AU], [0], marker="o", markersize=4, markeredgecolor="blue", markerfacecolor="blue")
text_e      = ax.text(AU,0,'Earth')

line_m,     = ax.plot([],[],'-g',lw=1)
point_m,    = ax.plot([1.666*AU], [0], marker="o", markersize=3, markeredgecolor="red", markerfacecolor="red")
text_m      = ax.text(1.666*AU,0,'Mars')

line_j,     = ax.plot([],[],'-g',lw=1)
point_j,    = ax.plot([AU], [0], marker="o", markersize=9, markeredgecolor="orange", markerfacecolor="orange")
text_j      = ax.text(AU,0,'Jupiter')

line_sat,     = ax.plot([],[],'-g',lw=1)
point_sat,    = ax.plot([AU], [0], marker="o", markersize=7, markeredgecolor="teal", markerfacecolor="teal")
text_sat      = ax.text(AU,0,'Saturn')

line_u,     = ax.plot([],[],'-g',lw=1)
point_u,    = ax.plot([AU], [0], marker="o", markersize=6, markeredgecolor="green", markerfacecolor="green")
text_u      = ax.text(AU,0,'Uranus')

line_n,     = ax.plot([],[],'-g',lw=1)
point_n,    = ax.plot([AU], [0], marker="o", markersize=6, markeredgecolor="black", markerfacecolor="black")
text_n      = ax.text(AU,0,'Neptune')

point_s,    = ax.plot([0], [0], marker="o", markersize=7, markeredgecolor="yellow", markerfacecolor="yellow")
text_s      = ax.text(0,0,'Sun')

merxdata,merydata = [],[]               # mercury track
venxdata,venydata = [],[]               # venus track
exdata,eydata = [],[]                   # earth track
sxdata,sydata = [],[]                   # sun track
mxdata,mydata = [],[]                   # mars track
jxdata,jydata = [],[]                   # jupiter track
satxdata,satydata = [],[]               # saturn track
uxdata,uydata = [],[]                   # uranus track
nxdata,nydata = [],[]                   # neptune track



def update(i):

    merxdata.append(xmerlist[i])
    merydata.append(ymerlist[i])

    venxdata.append(xvlist[i])
    venydata.append(yvlist[i])

    exdata.append(xelist[i])
    eydata.append(yelist[i])
    
    mxdata.append(xmlist[i])
    mydata.append(ymlist[i])

    jxdata.append(xjlist[i])
    jydata.append(yjlist[i])
    
    satxdata.append(xsatlist[i])
    satydata.append(ysatlist[i])
    
    uxdata.append(xulist[i])
    uydata.append(yulist[i])

    nxdata.append(xnlist[i])
    nydata.append(ynlist[i])

    line_mer.set_data(merxdata,merydata)
    point_mer.set_data(xmerlist[i],ymerlist[i])
    text_mer.set_position((xmerlist[i],ymerlist[i]))

    line_ven.set_data(venxdata,venydata)
    point_ven.set_data(xvlist[i],yvlist[i])
    text_ven.set_position((xvlist[i],yvlist[i]))

    line_e.set_data(exdata,eydata)
    point_e.set_data(xelist[i],yelist[i])
    text_e.set_position((xelist[i],yelist[i]))
    
    line_m.set_data(mxdata,mydata)
    point_m.set_data(xmlist[i],ymlist[i])
    text_m.set_position((xmlist[i],ymlist[i]))

    line_j.set_data(jxdata,jydata)
    point_j.set_data(xjlist[i],yjlist[i])
    text_j.set_position((xjlist[i],yjlist[i]))

    line_sat.set_data(satxdata,satydata)
    point_sat.set_data(xsatlist[i],ysatlist[i])
    text_sat.set_position((xsatlist[i],ysatlist[i]))

    line_u.set_data(uxdata,uydata)
    point_u.set_data(xulist[i],yulist[i])
    text_u.set_position((xulist[i],yulist[i]))

    line_n.set_data(nxdata,nydata)
    point_n.set_data(xnlist[i],ynlist[i])
    text_n.set_position((xnlist[i],ynlist[i]))
    
    point_s.set_data(xslist[i],yslist[i])
    text_s.set_position((xslist[i],yslist[i]))
    
    ax.axis('equal')
    ax.set_xlim(-30*AU,30*AU)
    ax.set_ylim(-30*AU,30*AU)
    plt.xlabel(r"Distance from Sun (meters)")
    plt.ylabel(r"Distance from Sun (meters)")
    plt.title(r"Orbits of the Planets in our Solar System")

    return    point_s,   text_s,
    line_mer, point_mer, text_mer, 
    line_ven, point_ven, text_ven
    line_e,   point_e,   text_e,
    line_m,   point_m,   text_m,
    line_j,   point_j,   text_j, 
    line_sat, point_sat, text_sat, 
    line_u,   point_u,   text_u, 
    line_n,   point_n,   text_n 

anim = animation.FuncAnimation(fig,func=update,frames=len(xelist),interval=1,blit=True)
plt.show()

```
